```python
from sympy import symbols
from sympy.plotting import plot
import matplotlib
import numpy as np

# 3a + 2b = 8
# 2a - b  = 3
x = np.array([[3, 2],[2, -1]])
y = np.array([8, 3])
z = np.linalg.solve(x, y)
```

```python
a, b = symbols('a b')
```

```python
annotations = [{
    "text": f"({z[0]}, {z[1]})",
    "xy": z + 0.5
}]
plot((2 * a) - 3, (8 - (3 * a)) / 2, annotations=annotations)
```

```python
from sympy.abc import x, y, z
from sympy.plotting import plot, plot3d
import sympy as sp
from sympy import solve, Eq

def plot_line(e):
    e = solve(e, y)
    plot(e[0])


def plot_plane(e):
    e = solve(e, z)
    plot3d(e[0])

def plot_eq(lhs, rhs):
    """ plots equation of form lhs = rhs. e.g to plot x+2y=3, call plot_eq(x+2*y, 3)"""
    plot(solve(Eq(lhs, rhs), y)[0])
```

```python
plot_eq(x + 2 * y, 3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.array([[8,-1],[4,9]])
b = np.array([9,7])
c = np.linalg.solve(a,b)

plt.figure()

# Set x-axis range
plt.xlim((-10,10))
# Set y-axis range
plt.ylim((-10,10))
# Draw lines to split quadrants
plt.plot([-10, 10], [0, 0], color='C0')
plt.plot([0, 0], [-10, 10], color='C0')

# Draw line 8a-b=9 => b=8a-9
x = np.linspace(-10, 10)
y = 8 * x - 9
plt.plot(x, y, color='C2')

# Draw line 4a+9b=7 => b=(7-4a)/9
y = (7 - 4*x) / 9
plt.plot(x, y, color='C2')

# Add solution
plt.scatter(c[0], c[1], marker='x', color='black')

# Annotate solution
plt.annotate('({:0.3f}, {:0.3f})'.format(c[0], c[1]), c+0.5)

plt.title('Quadrant plot')

plt.show()
```
